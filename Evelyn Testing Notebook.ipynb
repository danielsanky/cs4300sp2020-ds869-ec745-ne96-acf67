```python
import pandas as pd
from collections import Counter
```

```python
# Load music
music=pd.read_csv("young-people-survey/SpotifyFeatures.csv")
music=music.drop_duplicates("track_id")
music=music[music["popularity"].astype(float)>float(50)]
```

```python
genres = None # {'R&B', 'Alternative'}
genre_filter = lambda x: x['genre'] in genres if genres is not None else True
mask = music.apply(genre_filter, axis=1)
```

```python
music[mask]
```
